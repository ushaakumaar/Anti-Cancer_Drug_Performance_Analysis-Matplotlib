## Observations and Insights 

```python
# Dependencies and Setup
%matplotlib notebook

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
from scipy.stats import linregress

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_df = pd.merge(mouse_metadata, study_results, on="Mouse ID", how="outer")

# Display the data table for preview
mouse_df
```

```python
# Checking the number of mice.
number_of_mice = len(mouse_df["Mouse ID"].unique())
number_of_mice
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
dup_mouse_df = mouse_df[mouse_df.duplicated(["Mouse ID","Timepoint"], keep=False)]
dup_mouse_df
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouse_df = mouse_df.loc[mouse_df["Mouse ID"] != "g989"]
cleaned_mouse_df
```

```python
# Checking the number of mice in the clean DataFrame.
number_of_mice_now = len(cleaned_mouse_df["Mouse ID"].unique())
number_of_mice_now
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method is the most straighforward, creating multiple series and putting them all together at the end.

# Calculate Average of Tumor Volume per Drug Regimen
mean_stats = cleaned_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].mean()

# Calculate Median of Tumor Volume per Drug Regimen
median_stats = cleaned_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].median()

# Calculate Variance of Tumor Volume per Drug Regimen
var_stats = cleaned_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].var()

# Calculate Standard Deviation of Tumor Volume per Drug Regimen
std_stats = cleaned_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].std()

# Calculate SEM of Tumor Volume per Drug Regimen
sem_stats = cleaned_mouse_df.groupby("Drug Regimen")["Tumor Volume (mm3)"].sem()

# Create a DataFrame to hold above calculated statistics
stats_summary = pd.DataFrame({
    "mean" : mean_stats,
    "median": median_stats,
    "var" : var_stats,
    "std" : std_stats,
    "sem" : sem_stats
})
stats_summary
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# This method produces everything in a single groupby function
summary_stats = cleaned_mouse_df.groupby("Drug Regimen").agg({
    "Tumor Volume (mm3)" : ["mean", "median","var","std","sem"]
    })
summary_stats
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_timepoints = cleaned_mouse_df.groupby("Mouse ID")["Timepoint"].max()

# Merge this group df with the original dataframe to get the tumor volume at the last timepoint
new_merged_df = pd.merge(cleaned_mouse_df, last_timepoints,on=("Mouse ID","Timepoint"), how="inner")
new_merged_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatment_list = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 

#Loop through each treatment
for treatment in treatment_list:

    # Locate the rows which contain mice on each drug and get the tumor volumes
    tumor_vol = new_merged_df.loc[(new_merged_df["Drug Regimen"] == treatment)]["Tumor Volume (mm3)"]
    
    # add subset 
    tumor_vol_data.append(tumor_vol)
    
    # Determine outliers for treatment using upper and lower bounds
    quartiles = tumor_vol.quantile([.25,.75])
    
    # Determine 25th Quartile and 75th Quartile for treatment
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    
    # Calculate the IQR for treatment
    iqr = upperq - lowerq
    
    # Calculate the lower and upper bounds for treatment
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    # Find potential outliers for treatment using the lower and upper bounds
    outliers = tumor_vol.loc[(tumor_vol < lower_bound) | (tumor_vol > upper_bound)]
    if len(outliers) > 0:
        print(f"There is {len(outliers)} Outlier(s) for {treatment}")
        for outlier in outliers:
            print(outlier)
```

```python
# Generate a box plot of the final tumor volume of each mouse across four regimens of interest

# Set unique color for the outliers
red_dot = dict(markerfacecolor='r', marker='o')

# Generate boxplot 
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volumes / Treatment - Box Plot')
ax1.boxplot(tumor_vol_data, flierprops=red_dot, labels = treatment_list)

# Set x and y labels
plt.xlabel("Treatments")
plt.ylabel("Final Tumor Volume (mm3)")

# Save and Display the box plot
plt.tight_layout()
plt.savefig("Images/boxplot.png")
plt.show()
```
